```python
#Squares of Numbers from 1 to 25

def print_squares_1_to_25():
    squares = []
    
    for i in range(1, 26):
        squares.append(i ** 2)
        
    print("Squares from 1 to 25:", squares)

print_squares_1_to_25()
```

```python
#Find the Highest Number

def highest(n1, n2, n3):
    
    if n1 > n2 and n1 > n3:
        return n1
    
    elif n2 > n1 and n2 > n3:
        return n2
    
    elif n3 > n1 and n3 > n2:
        return n3
    
    elif n1 == n2 and n2 == n3:
        return "equal"


num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

result = highest(num1, num2, num3)

print("Result:", result)
```

```python
#Expand Binomial Expression

from sympy import symbols, expand

x, y = symbols('x y')

expr = (x + y)**2

expanded = expand(expr)

print("Expanded form:", expanded)
```

```python
#Solve System of Equations

from sympy import symbols, Eq, solve

x, y = symbols('x y')

eq1 = Eq(3*x + 2*y, 15)
eq2 = Eq(x - y, 2)

solution = solve((eq1, eq2), (x, y))

print("Solution:", solution)
```

```python
#Substitute Values in Expression

from sympy import symbols

x, y = symbols('x y')

expr = x**2 + 2*y

result = expr.subs({x: 4, y: 7})

print("Result after substitution:", result)
```
