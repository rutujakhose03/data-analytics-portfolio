```python
#Prime Number Program

a = int(input("enter any number: "))

if a <= 1:
    print(a, "is not a prime number")
else:
    for i in range(2, a):
        if a % i == 0:
            print(a, "is not a prime number")
            break
    else:
        print(a, "is a prime number")
```

```python
#Print @
#      ##
#      @@@
#      ####
#      @@@@@
#      ######

n = 6
for i in range(1, n+1):
    if i % 2 != 0:
        print("@" * i)
    else:
        print("#" * i)
```

```python
#sum of numbers from 1 to 10

i = 1
sum = 0
while (i <= 10):
    sum += i
    i += 1
print("sum is:", sum)
```
