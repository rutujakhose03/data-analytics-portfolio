```python
#Volume of a Cuboid

length = 7
width = 8
height = 9

volume = length * width * height

print("Volume of cuboid:-", volume, "cubic units")
```

```python
#Calculate Speed

distance = 350  # km
time = 6        # hours

speed = distance / time

print("Speed =", speed, "km/hr")
```

```python
#Currency Converter

inr = float(input("Enter amount in INR"))

usd = inr / 83.2

print("Amount in USD:", round(usd, 2))
```

```python
#Convert Days to Weeks and Days

days = int(input("Enter number of days"))

weeks = days // 7
extra_days = days % 7

print(f"{days} days = {weeks} week(s) and {extra_days} day(s)")
```
