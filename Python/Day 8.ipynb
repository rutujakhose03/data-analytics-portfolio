```python
#Age Caegory
age = int(input("Age"))
if(age < 0):
    print("Invalid age")
elif(age <= 12):
    print("Child")
elif(age <= 17):
    print("Teen")
elif(age <= 64):
    print("Adult")
else:
    print("Senior")
```

```python
#triangle type
a = float(input("Side a: "))
b = float(input("Side b: "))
c = float(input("Side c: "))
if(a + b > c and a + c > b and b + c > a):
    if(a == b == c):
        print("Equilateral triangle")
    elif a == b or b == c or a == c:
        print("Isosceles triangle")
    else:
        print("Scalene triangle")
else:
    print("Not a valid triangle")
```

```python
#ATM stimulation
balance = 20000
withdraw = int(input("Enter amount to withdraw"))
if(withdraw <= balance):
    balance -= withdraw
    print("Transaction successful. Remaining balance:", balance)
else:
    print("Insufficient balance!")
```

```python
#garding system
score = int(input("Enter your score (0–100)"))
if(score >= 90):
    print("Grade: I")
elif(score >= 80):
    print("Grade: II")
elif(score >= 70):
    print("Grade: III")
elif(score >= 60):
    print("Grade: IV")
else:
    print("Grade: V")
```
