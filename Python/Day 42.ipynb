```python
#Currency Note Counter

amount = int(input("Enter withdrawal amount:- ₹"))

notes = [2000, 500, 200, 100, 50, 20, 10]

for note in notes:
    count = amount // note
    if count:
        print(f"{note} x {count}")
    amount %= note
```
