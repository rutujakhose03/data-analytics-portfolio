```python
#Division of Complex Number

c1 = complex(5,6)
c2 = complex(2,-3)
div = c1/c2
if (div!=0):
    print("Division :", div)
else:
    print("result")
```

```python
#Conjugate of Complex Number (Direct Method)

c = complex(2,-3)
print("conjugate of :",c,"is", c.conjugate())
```

```python
#Conjugate of Complex Number (Using Variable)

c = complex(5,6)
conj = c.conjugate()
print("conjugate :", conj)
```

```python
#Modulus of Complex Number

c = complex(5,6)
modulus = abs(c)
print("complex number :",c,"is",modulus)
```

```python
#Number Pattern Program

z = 1
for i in range(1,11):
    for j in range(1,i+1):
        print(z,end=" ")
        z += 1
    print()
```
