```python
# write complex number
c1 = complex(5, 6)
c2 = complex(2, -3)

if c1 == c2:
    print("Both complex numbers are equal:", c1)
else:
    print("Complex numbers are different")
    print("c1 =", c1)
    print("c2 =", c2)
```

```python
# write addition of complex number
c1 = complex(5, 6)
c2 = complex(2, -3)

add = c1 + c2

if add != 0:
    print("Addition:", add)
else:
    print("Result is zero")
```

```python
# subtraction of complex number
c1 = complex(5, 6)
c2 = complex(2, -3)

sub = c1 - c2

if sub != 0:
    print("Subtraction:", sub)
else:
    print("Result is zero")
```

```python
# Multiplication of complex number
c1 = complex(5, 6)
c2 = complex(2, -3)

mul = c1 * c2

if mul != 0:
    print("Multiplication:", mul)
else:
    print("Result is zero")
```
