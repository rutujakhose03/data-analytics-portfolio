```python
# Expand (a - b)^3

from sympy import symbols, expand

a, b = symbols('a b')

expr = (a - b)**3

expanded = expand(expr)

print("Expanded form:", expanded)
```

```python
# Solve equations

from sympy import symbols, Eq, solve

x, y = symbols('x y')

eq1 = Eq(x + y, 16)
eq2 = Eq(x - y, 8)

solution = solve((eq1, eq2), (x, y))

print("Solution:", solution)
```

```python
#Invoke Function After Specific Time

import time
import math

def delayed_sqrt(numbers, delay_ms):

    delay_seconds = delay_ms / 1000.0

    print("Square root after specific milliseconds:")

    for num in numbers:
        time.sleep(delay_seconds)
        print(math.sqrt(num))


numbers = [56, 150, 26789]
delay_ms = 500

delayed_sqrt(numbers, delay_ms)
```

```python
#Perimeter of a Square

def perimeter_of_square(side):
    peri = 4 * side
    print("Perimeter of square:", peri)

side_length = float(input("Enter the side length of the square: "))

perimeter_of_square(side_length)
```

```python
#Basic Algebraic Expression (2x + 3)

def evaluate_expression(x):
    return 2 * x + 3

x = float(input("Enter the value of x: "))

result = evaluate_expression(x)

print("Result of 2x + 3 is:", result)
```
