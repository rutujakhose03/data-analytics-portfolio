```python
#Display All Factors of a Number

num = int(input("Enter a number"))

print("Factors of", num, "are:", end=" ")

for i in range(1, num + 1):
    if num % i == 0:
        print(i, end=" ")
```

```python
#Convert Days to Years and Remaining Days

days = int(input("Enter number of days"))

years = days // 365
remaining_days = days % 365

print(f"{days} days = {years} year(s) and {remaining_days} day(s)")
```

```python
#Area of a Triangle Using Heron’s Formula

a = 5
b = 6
c = 7

s = (a + b + c) / 2

area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Area of triangle:", area)
```

```python
#Check Whether a Number is Triangular

num = 18

i = 1
sum = 0

while sum < num:
    sum += i
    i += 1

print("Triangular Number" if sum == num else "Not a Triangular Number")
```

```python
#Convert Temperature from Celsius to Kelvin

celsius = 45

kelvin = celsius + 273.15

print("Temperature in Kelvin:", kelvin)
```
