```python
# Create a tuple of vegetables
vegetables = ("Tomato", "Bottle Gourd", "Potato", "Pea", "Green Chili", "Cucumber", "Onion", "Cauliflower", "Bitter Gourd")

print(vegetables)
```

```python
# Accessing Tuple Elements
fruits = ("apple", "banana", "orange", "grapes", "mango", "strawberry", "pineapple", "watermelon")

print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])
print(fruits[4])
```

```python
# Concatenate and Repeat Tuples
tuple1 = (1, 2, 3, 4, 5, 6)
tuple2 = (7, 8, 9, 10)

result = tuple1 + tuple2
repeat = tuple1 * 2

print("Concatenated:", result)
print("Repeated:", repeat)
```

```python
# Count and Index Methods
data = (1, 2, 3, 2, 4, 2, 5)

print(data.count(2))
print(data.index(3))
```

```python
# Convert Tuple to List and Back
tuple1 = (10, 20, 30, 40, 50, 60, 70)

list1 = list(tuple1)   # Convert to list
list1.append(40)       # Add new element

tuple2 = tuple(list1)  # Convert back to tuple

print("Modified tuple:", tuple2)
```
